## Essential Functionality

### Reindexing

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj
```

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])#重排，不存在引入缺失值
obj2
```

```python
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'],fill_value=0)#用0替代缺失值
obj2
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3
```

```python
obj3.reindex(range(6), method='ffill')#重新索引可能需要做一些插值处理，method可以做到。
```

```python
frame = pd.DataFrame(np.arange(9).reshape((3, 3)),
                     index=['a', 'c', 'd'],
                     columns=['Ohio', 'Texas', 'California'])
frame
```

```python
frame2 = frame.reindex(['a', 'b', 'c', 'd'])
frame2
```

```python
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states)#使用columns即可重新索引列
```

```python
frame
```

```python
frame2=frame.reindex(index=['a', 'b', 'c', 'd'],columns = states.sort(),method = 'ffill')#对行和列同时索引
frame2
```

```python
# Assign tie values the maximum rank in the group
obj.rank(ascending=False, method='max')#降序排名，使用整个分组的最大排名
```

### Correlation and Covariance

conda install pandas-datareader

```python
import pandas_datareader.data as web
import fix_yahoo_finance as fy
fy.pdr_override()
```

```python
all_data = {ticker: web.get_data_yahoo(ticker,'2000-01-01','2010-01-01')
            for ticker in ['AAPL', 'IBM', 'MSFT', 'GOOG']} #获取2000/01/01到2010/01/01四只股票的数据
```

```python
price = pd.DataFrame({ticker: data['Adj Close']
                     for ticker, data in all_data.items()})#价格
volume = pd.DataFrame({ticker: data['Volume']
                      for ticker, data in all_data.items()})#成交量
```

```python
returns = price.pct_change()#计算价格的百分数变化
returns.tail()
```

```python
returns['MSFT'].corr(returns['IBM'])#corr方法计算两个Series中重叠的、非NA的、按索引对齐的值得相关系数
```

```python
returns['MSFT'].cov(returns['IBM'])#计算协方差
```

```python
returns.MSFT.corr(returns.IBM)
```

```python
returns.corr()#DataFrame的corr和cov方法将以DataFrame的形式返回完整的相关系数或协方差矩阵
```

```python
returns.cov()
```

```python
#利用DataFrame的corrwith方法，可以计算其列或行跟另一个Series或DataFrame之间的相关系数
#传入一个Series将会返回一个相关系数值Series
returns.corrwith(returns.IBM)
```

```python
#传入一个DataFrame则会计算按列名配对的相关系数
#这里计算百分比变化与成交量的相关系数
returns.corrwith(volume)
```
